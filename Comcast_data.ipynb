```python
import pandas as pd
```

```python
path = r"Comcast_telecom_complaints_data.csv"  # paste the path here
df_complaints = pd.read_csv(path)
df_complaints.head()
```

```python
df_complaints.info()
```

```python
df_complaints["Date_month_year"] = pd.to_datetime(
    df_complaints["Date_month_year"], format="%d-%b-%y"
)
df_complaints.head()
```

```python
df_complaints.info()
```

```python
df_complaints["Time"] = pd.to_datetime(
    df_complaints["Time"], format="mixed"
).dt.strftime("%H:%M:%S")
df_complaints["Time"] = pd.to_timedelta(df_complaints["Time"])
df_complaints.head()
```

```python
df_complaints["date_index"] = df_complaints["Date_month_year"] + df_complaints["Time"]
df_complaints.head()
```

```python
df_complaints = df_complaints.drop(["Date_month_year", "Time"], axis=1)
df_complaints.head()
```

```python
df_complaints = df_complaints.set_index(df_complaints["date_index"])
df_complaints = df_complaints.drop(['date_index'], axis=1)
df_complaints.head()
```
